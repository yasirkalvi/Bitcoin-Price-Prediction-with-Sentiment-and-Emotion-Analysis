```python
#Cleaning

from nltk.tokenize import TweetTokenizer
nltk.download('stopwords')
stopword = nltk.corpus.stopwords.words('english')

##Functio\n to clean the tweets:
def cleanTxt(text):
    text = re.sub(r'@[A-Za-z0-9]+', '', text) #Remvoded @mentions

    #text = re.sub(r'#', '', text) #Removing the '#' symbol

    text = re.sub(r'RT[\s]+', '', text) #Removing RT

    text = re.sub(r'https?:\/\/\S+', '', text) #Remove the hyper link

    #text = text.lower() #converting text to lower case

    text = re.sub("[0-9]","", text) #removing digits
    
    text = re.sub(r'[^\w\s]',"", text) #removing punctuation

    #text = re.sub("[^a-z0-9]"," ", text) #removing all non-alphanumeric characters
    
    pattern = re.compile(r'\b(' + r'|'.join(stopword) + r')\b\s*')
    text = pattern.sub('', text) #Remove stop words

    #text = [word for word in text if word not in stopword]

    ##pos tagging
    return text
```

```python
#Cleaning the text
df['Tweets']= df['Tweets'].apply(cleanTxt)

#Show the cleaned text
df['Tweets']
```

```python
def tokenizer(text):

    text = TweetTokenizer(   #using tweet tokenizer from nltk library

        preserve_case=False, #preserve_case lower-cases the letters

        strip_handles=True, #strip_handles removes the twitter-handles

        reduce_len=True).tokenize(text) #reduce_len limits the number of times a character can be repeated to three
    
    return text
```

```python
#Tokenising the text
df['Tweets_tokens']= df['Tweets'].apply(tokenizer)

#Show the cleaned text
df['Tweets_tokens']
```

```python
def lemmatizer(tokens):

    from nltk.corpus import wordnet

    def get_wordnet_pos(treebank_tag):

        if treebank_tag.startswith('J'):
            return wordnet.ADJ
        elif treebank_tag.startswith('V'):
            return wordnet.VERB
        elif treebank_tag.startswith('N'):
            return wordnet.NOUN
        elif treebank_tag.startswith('R'):
            return wordnet.ADV
        else:
            return None # for easy if-statement 



    from nltk.stem.wordnet import WordNetLemmatizer

    lemmatizer = WordNetLemmatizer()

    l=[]
    for token, tag in pos_tag(tokens):

        wntag = get_wordnet_pos(tag)
        
        
        if wntag is None:# not supply tag in case of None
            l.append(lemmatizer.lemmatize(token))
        else:
            l.append(lemmatizer.lemmatize(token, pos=wntag))
            
    
    return l
            
```

```python
df['Lemma_Tweets'] = df['Tweets_tokens'].apply(lambda x:lemmatizer(x))

print(df['Lemma_Tweets'])
print(df['Tweets'])
```

```python
def get_emotions(text, dataframe):
    import operator
    """
    Function that will take a text at return a dataframe of emotions (excluding positive and negative)
    Params:
        text (str): The tweet of interest
        dataframe (Pandas DataFrame): The emotions dataframe to add values
        
    Return:
        Updated dataframe with the new row of interest
    """
    nrc_text = NRCLex(text)
    nrc_dict = nrc_text.affect_frequencies
    nrc_dict.pop('positive', None)
    nrc_dict.pop('negative', None)
    nrc_dict.pop('anticip', None)
    nrc_max_emotion = max(nrc_dict.items(), key=operator.itemgetter(1))[0]
    nrc_dict["max_emotion"] = nrc_max_emotion
    dataframe = dataframe.append(nrc_dict, ignore_index= True)
    return dataframe
```
